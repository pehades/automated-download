```python
import requests
import os.path

from bs4 import BeautifulSoup
```

We want to enter a site and download a fraction of the pdf (or other extension files) included inside. For example, we follow the Stanford Course Machine Learning by Andrew Ng, and want to download alll the lecture notes automatically.

```python
url = 'https://see.stanford.edu/Course/CS229'

r = requests.get(url)
soup = BeautifulSoup(r.content, 'html5lib')

for link in soup.findAll('a'):
    print(link['href'])
```

```python
#We see that the notes are of the form '/materials/aimlcs229/cs229-notesXX.pdf'.

#Therefore we access then by their unique title '/materials/aimlcs229/cs229-notes' 

generic_dom = '/materials/aimlcs229/cs229-notes'

# Length of generic_dom is 32

for link in soup.findAll('a'):    
    if link['href'][:32] == generic_dom:
        title = link['href'][27:]
        f = requests.get(url + link['href'])
        if not os.path.isfile(title):                     #this dumps the file into current directory
            with open(title, 'wb') as file:
                for chunk in f.iter_content(chunk_size = 1024*1024):
                    if chunk:
                        file.write(chunk)
```

Of course this can happen with every webpage that contains pdf files. Just take care of the url and when iterating through the links, find the specific form they have, or the specific range these files are.
